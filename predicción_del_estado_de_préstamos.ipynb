Se usa un modelo como Random Forest para calcular la importancia de las características porque este tipo de algoritmo puede medir qué tan influyente es cada variable en la predicción del resultado.

Cada árbol en el bosque selecciona características al azar y mide cómo afectan la predicción.
Se calcula una métrica de importancia basada en la reducción de impureza (Gini o entropía) o el impacto en las decisiones del modelo.

```python
def evaluate_classification_models(X_train, y_train, X_test, y_test, models):
    """
    Evalúa una lista de modelos, almacena los resultados y devuelve un DataFrame para comparación.

    Args:
        X_train (pd.DataFrame): Características de entrenamiento.
        y_train (pd.Series): Variable objetivo de entrenamiento.
        X_test (pd.DataFrame): Características de prueba.
        y_test (pd.Series): Variable objetivo de prueba.
        models (list): Lista de modelos de clasificación de scikit-learn.

    Returns:
        pd.DataFrame: DataFrame con los nombres de los modelos y sus métricas de evaluación.
    """

    model_results = []
    trained_models = {}

    for model in models:
        model.fit(X_train, y_train)  # Entrenar el modelo
        trained_models[model.__class__.__name__] = model  # Guardar el modelo entrenado con su nombre

        start_time = time.time()  # Registrar el tiempo de inicio
        prediction = model.predict(X_test)  # Realizar predicciones
        inference_time = time.time() - start_time  # Calcular el tiempo de inferencia

        model_results.append({
            "Nombre modelo": model.__class__.__name__,  # Nombre del modelo
            "Accuracy": accuracy_score(y_test, prediction) * 100,  # Precisión
            "ROC_AUC": roc_auc_score(y_test, prediction),  # Área bajo la curva ROC
            "F1_Score": f1_score(y_test, prediction),  # Puntuación F1
            "Precision": precision_score(y_test, prediction),  # Precisión
            "Recall": recall_score(y_test, prediction),  # Sensibilidad o recall
            "Inference Time (ms)": inference_time * 1000  # Tiempo de inferencia en milisegundos
        })

    models_df = pd.DataFrame(model_results)  # Convertir resultados en DataFrame
    models_df = models_df.set_index('Nombre modelo')  # Establecer el índice en el nombre del modelo
    models_df = models_df.sort_values("F1_Score", ascending=False)  # Ordenar por mejor F1 Score

    return models_df, trained_models  # Retornar el DataFrame con métricas y los modelos entrenados
```

## Evaluacion de modelos

```python
def evaluate_classification_metrics(y_true, y_pred, target_names=None, display=True):
    """
    Evalúa un modelo de clasificación generando una matriz de confusión,
    un reporte de clasificación y opcionalmente una visualización de la matriz de confusión.

    Args:
        y_true (pd.Series): Etiquetas reales de la clasificación.
        y_pred (pd.Series): Etiquetas predichas por el modelo.
        target_names (list, opcional): Nombres de las clases para mejorar la legibilidad de la matriz de confusión.
        display (bool, opcional): Si es True, muestra la matriz de confusión gráficamente.

    Returns:
        dict: Un diccionario con la matriz de confusión, el reporte de clasificación y los nombres de las clases (si se proporcionan).
    """

    # Convertimos y_true y y_pred a arrays de NumPy para asegurar compatibilidad con métricas de sklearn
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    # Calculamos la matriz de confusión
    cm = confusion_matrix(y_true, y_pred)

    # Generamos el reporte de clasificación con métricas como precisión, recall y F1-score
    report = classification_report(y_true, y_pred, target_names=target_names)

    # Si display es True, mostramos la matriz de confusión visualmente
    if display:
        ConfusionMatrixDisplay(cm, display_labels=target_names).plot()
        plt.show()  # Mostramos la matriz de confusión

    # Retornamos los resultados en un diccionario para fácil acceso
    evaluation_results = {
        "Confusion Matrix": cm,  # Matriz de confusión en formato numérico
        "Classification Report": report,  # Reporte detallado del desempeño del modelo
        "Target Names": target_names,  # Guardamos los nombres de las clases si se proporcionaron
    }

    return evaluation_results
```
